# Effect of the Student grade according to Different Variables

## Investigation Overview

In this investigation, I wanted to look on how different conditions affect our students grade and how to apply them. Which can give us a better results for our children, kids, students ..etc. The main focus in this investigation is on the four conditions that affect our students for a better grades in "Math", "Writing", and "Reading" test.

## Dataset Overview

The data consisted of 1,000 student with 8 columns. The attributes that affect the student **Race/Ethnicity**, **Gender**, **Parental level**, **Lunch**, and **Preperation Test** and the outcome of the student grades in "Math", "Reading",and "Writing" test. No data has been removed as their was no null values or duplicates

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
df=pd.read_csv('./StudentsPerformance.csv')
df.head(10)
```

```python
df.rename(columns={'race/ethnicity': 'ethnicity','parental level of education':'parent_education','test preparation course':'prep_course','math score':'math_score','reading score':'reading_score','writing score':'writing_score'},inplace=True)
```

```python
df.isna().sum()
```

# Distrubution of our students Tests
In the math test our grades are normally distrubuted and can be clarify not many conditions affect their grades. On the other hand, the **Writing** and **Reading** test has a more spikey values make us wonder why their is **sudden fall** in grades and **how to fix** these problems?.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
# Math test
sb.histplot(df.math_score,ax=axes[0],bins=50);
axes[0].title.set_text('Math Test ')
axes[0].set_xlabel('Grades')
axes[0].set_ylabel('Frequency');
# Writing test
sb.histplot(df['writing_score'], ax=axes[1],bins=50);
axes[1].title.set_text('Writing Test ')
axes[1].set_xlabel('Grades')
axes[1].set_ylabel('Frequency');
# # Reading Test
sb.histplot(df['reading_score'], ax=axes[2],bins=50);
axes[2].title.set_text('Reading Test ')
axes[2].set_xlabel('Grades')
axes[2].set_ylabel('Frequency');
fig.tight_layout()
```

# Effect of taking a preparation test.
After the students have taken the preparation test it shifted at least 20% better grades. You can look at the differnce of the points where started.
 - **Prep Course** starts from 30% at least
 - **No Prep Course** starts from aproximately 20%.

```python
#Math Test
had_test_m=df.query('prep_course == "completed"')['math_score'].value_counts()
had_no_test_m=df.query('prep_course == "none"')['math_score'].value_counts()
#Writing Test
had_test_w=df.query('prep_course == "completed"')['writing_score'].value_counts()
had_no_test_w=df.query('prep_course == "none"')['writing_score'].value_counts()
#Reading Test
had_test_r=df.query('prep_course == "completed"')['reading_score'].value_counts()
had_no_test_r=df.query('prep_course == "none"')['reading_score'].value_counts()
print('Math Test with prep_test median = ',had_test_m.median()*10,'  Math Test without prep_test median = ',had_no_test_m.median()*10)
print('Writing Test with prep_test median = ',had_test_w.median()*10,'  Writing Test without prep_test median = ',had_no_test_w.median()*10)
print('Reading Test with prep_test median = ',had_test_r.median()*10,'  Reading Test without prep_test median = ',had_no_test_r.median()*10)
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1, figsize = [10,10])
#Brown
color1=(0.69411766529083252, 0.3490196168422699, 0.15686275064945221, 1.0)
#Light Blue
color2=(0.65098041296005249, 0.80784314870834351, 0.89019608497619629, 1.0)

#Math Test
ax[0].scatter(x=had_test_m.index, y=had_test_m,color=color1,label='Completed')
#Didn't complete Preparation Course
ax[0].scatter(x=had_no_test_m.index, y=had_no_test_m,color=color2,label='None')
ax[0].title.set_text('Math test -> Preperation test effect')
ax[0].set_xlabel('Grades')
ax[0].set_ylabel('Frequency')
ax[0].set_yticks(np.arange(0, 30, step=5))
ax[0].legend(title='Preperation Course')

#Writing Test
ax[1].scatter(x=had_test_w.index, y=had_test_w,color=color1,label='Completed')
#Didn't complete Preparation Course
ax[1].scatter(x=had_no_test_w.index, y=had_no_test_w,color=color2,label='None')
ax[1].title.set_text('Writing test -> Preperation test effect')
ax[1].set_xlabel('Grades')
ax[1].set_ylabel('Frequency')
ax[1].set_yticks(np.arange(0,30, step=5))
ax[1].legend(title='Preperation Course')

#Reading Test
ax[2].scatter(x=had_test_r.index, y=had_test_r,color=color1,label='Completed')
#Didn't complete Preparation Course
ax[2].scatter(x=had_no_test_r.index, y=had_no_test_r,color=color2,label='None')
ax[2].title.set_text('Reading test -> Preperation test effect')
ax[2].set_xlabel('Grades')
ax[2].set_ylabel('Frequency')
ax[2].set_yticks(np.arange(0, 30, step=5))
ax[2].legend(title='Preperation Course')

fig.tight_layout()
```

# Effect of gender on the student grades.
According to our figure below:-
 - Math test:- **Male** is better than **Female**.
 - Reading and Writing:- **Females** are much better than males after 70% as we can see in the figure below the frequancy difference is big.
 - Giving that male's needs more care and assiting in their reading and writing skills.

```python
fig, ax = plt.subplots(nrows=1,ncols=3, figsize = [20,7])

#Brown
color1=(0.69411766529083252, 0.3490196168422699, 0.15686275064945221, 1.0)
#Light Blue
color2=(0.65098041296005249, 0.80784314870834351, 0.89019608497619629, 1.0)

#Math
sb.histplot(data = df, x = 'math_score', hue = 'gender',ax=ax[0])
ax[0].set_title('Math Test -> Comparing Gender')
ax[0].set_xlabel('Grades')
ax[0].set_ylabel('Frequency')
ax[0].set_yticks(np.arange(0, 70, step=10))

#Writing
sb.histplot(data = df, x = 'writing_score', hue = 'gender',ax=ax[1])
ax[1].set_title('Writing Test -> Comparing Gender')
ax[1].set_xlabel('Grades')
ax[1].set_ylabel('Frequency')

#Reading
sb.histplot(data = df, x = 'reading_score', hue = 'gender',ax=ax[2])
ax[2].set_title('Reading Test -> Comparing Gender')
ax[2].set_xlabel('Grades')
ax[2].set_ylabel('Frequency')
ax[2].set_yticks(np.arange(0, 70, step=10))

fig.tight_layout()
```

# Let us see the affect of ethinicty/race on grades

There are 5 groups of ethinicity as shown in the figure below, we can say that group A has the lowest grades in the 3 exams. But, we saw a major difference when they taken a preperation test. A major improvement in all test giving us the hint that group A needs our care and giving more attention on their learning syllabus that need to be improved. We conclude that group A has a great potential students, but need a extra care on their school learning system that we need to figure it out.

```python

plt.figure(figsize=(10, 15))
# Types of sedan cars
type_groups = ['group A', 'group B', 'group C', 'group D', 'group E']

# Returns the types for sedan_classes with the categories and orderedness
# Refer - https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.api.types.CategoricalDtype.html
vgroupes = pd.api.types.CategoricalDtype(ordered=True, categories=type_groups)

# Use pandas.astype() to convert the "VClass" column from a plain object type into an ordered categorical type 
df.ethnicity = df.ethnicity.astype(vgroupes);
base_color = sb.color_palette()[0]


plt.subplot(3, 1, 1)
sb.violinplot(data=df, x='ethnicity', y='math_score',hue='prep_course',inner='quartile',color = base_color);
plt.xticks(rotation=45);
plt.title('Math Test -> Ethnicity effect')

plt.subplot(3, 1, 2)
base_color = sb.color_palette()[1]
sb.violinplot(data=df, x='ethnicity', y='writing_score',hue='prep_course', inner='quartile', color = base_color);
plt.xticks(rotation=45);
plt.title('Writing Test -> Ethnicity effect')

plt.subplot(3, 1, 3)
base_color = sb.color_palette()[2]
sb.violinplot(data=df, x='ethnicity', y='reading_score',hue='prep_course', inner='quartile', color = base_color);
plt.xticks(rotation=45);
plt.title('Reading Test -> Ethnicity effect')

plt.tight_layout()
```
